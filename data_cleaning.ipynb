## Handling Missing Values
### Option 1 : Imputation (using median for numerical columns)
### Option 2 : Deletion (drop rows with missing values)

#### Issues currently : Experience col which is our target variable it has high missing values, we need to resolve that and deletion is not an option 

```python
import numpy as np

# ---------------------
# 1. Imputation Section
# ---------------------

# Convert Experience to numeric (invalid entries become NaN)
df['Experience'] = pd.to_numeric(df['Experience'], errors='coerce')

# Impute missing Experience with median
df['Experience'] = df['Experience'].fillna(df['Experience'].median())

# Impute missing Skills and Description
df['Skills'] = df['Skills'].fillna('')
df['Description'] = df['Description'].fillna('')
df['Title'] = df['Title'].fillna('Unknown')

# -----------------------------
# 2. Feature Engineering Section
# -----------------------------

# Skill count
df['skill_count'] = df['Skills'].apply(lambda x: len([s for s in x.split(',') if s.strip()]))

# Job level inferred from Title
def infer_level(title):
    title = title.lower()
    if any(w in title for w in ['entry', 'graduate', 'junior']):
        return 'entry'
    elif any(w in title for w in ['mid', 'intermediate']):
        return 'mid'
    elif any(w in title for w in ['senior', 'lead', 'principal']):
        return 'senior'
    else:
        return 'unknown'

df['job_level'] = df['Title'].apply(infer_level)
job_level_dummies = pd.get_dummies(df['job_level'], prefix='level')

# Description length
df['desc_length'] = df['Description'].apply(len)

# Log-transformed Experience
df['Experience_log'] = np.log1p(df['Experience'])

# ------------------------------------
# 3. Correlation Matrix and Heatmap
# ------------------------------------

# Combine engineered features for correlation analysis
df_engineered = pd.concat([
    df[['Experience', 'Experience_log', 'skill_count', 'desc_length']],
    job_level_dummies
], axis=1)

# Plot correlation heatmap
plt.figure(figsize=(10, 6))
corr_matrix = df_engineered.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap with Experience and Engineered Features")
plt.tight_layout()
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler

# Select only numeric features (after feature engineering)
numeric_features = df_engineered.select_dtypes(include=[np.number])

# Drop any remaining NaNs (should be safe after imputation)
X = numeric_features.dropna()

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]

# Print VIF results
print("Variance Inflation Factor (VIF):")
print(vif_data)

# Experience and Experience_log: vif occured 3.66, 
# which is moderate level of correlation,
# to avoid redundancy we will drop Experience column
if 'Experience' in df_engineered.columns:
    df_engineered.drop(columns=['Experience'], inplace=True)

# Plot correlation heatmap
plt.figure(figsize=(10, 6))
corr_matrix = df_engineered.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap with Experience and Engineered Features")
plt.tight_layout()
plt.show()

df_engineered.head()
```
